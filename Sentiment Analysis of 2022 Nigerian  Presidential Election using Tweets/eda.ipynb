```python
df_obi["user_location"] = df_obi["user_location"].replace(['Lagos, Nigeria','Lagos Nigeria',"Lagos,Nigeria"],'Lagos')
df_obi["user_location"] = df_obi["user_location"].replace(['Federal Capital Territory, Nig','Abuja, Nigeria'],'Abuja')
df_obi["user_location"] = df_obi["user_location"].replace(['Port Harcourt, Nigeria','Rivers, Nigeria'],'Port Harcourt')
df_obi["user_location"] = df_obi["user_location"].replace(['Ibadan, Nigeria',],'Ibadan')
df_obi["user_location"] = df_obi["user_location"].replace(['Kaduna, Nigeria',],'Kaduna')
df_obi["user_location"] = df_obi["user_location"].replace(['Owerri, Nigeria',],'Owerri')
df_obi["Sentiment"] = df_obi["Sentiment"].replace([0],'Negative')
df_obi["Sentiment"] = df_obi["Sentiment"].replace([1],'Positive')
```

```python
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Lagos, Nigeria ','Lagos Nigeria',"Lagos Nigeria","Lagos, Nigeria"],'Lagos')
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Federal Capital Territory, Nig','Abuja, Nigeria'],'Abuja')
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Port Harcourt, Nigeria','Rivers, Nigeria'],'Port Harcourt')
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Ibadan, Nigeria',],'Ibadan')
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Kaduna, Nigeria',],'Kaduna')
df_tinubu["user_location"] = df_tinubu["user_location"].replace(['Owerri, Nigeria',],'Owerri')
df_tinubu["Sentiment"] = df_tinubu["Sentiment"].replace([0],'Negative')
df_tinubu["Sentiment"] = df_tinubu["Sentiment"].replace([1],'Positive')
```

```python
df_atiku["user_location"] = df_atiku["user_location"].replace(['Lagos, Nigeria','Lagos Nigeria',"Lagos Nigeria","Lagos, Nigeria"],'Lagos')
df_atiku["user_location"] = df_atiku["user_location"].replace(['Federal Capital Territory, Nig','Abuja, Nigeria'],'Abuja')
df_atiku["user_location"] = df_atiku["user_location"].replace(['Port Harcourt, Nigeria'],'Port Harcourt')
df_atiku["user_location"] = df_atiku["user_location"].replace(['Ibadan, Nigeria',],'Ibadan')
df_atiku["user_location"] = df_atiku["user_location"].replace(['Kaduna, Nigeria',],'Kaduna')
df_atiku["user_location"] = df_atiku["user_location"].replace(['Owerri, Nigeria',],'Owerri')
df_atiku["Sentiment"] = df_atiku["Sentiment"].replace([0],'Negative')
df_atiku["Sentiment"] = df_atiku["Sentiment"].replace([1],'Positive')
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14, 5))

sns.kdeplot(df_obi['user_favourites'], label = 'Obi', shade = True, color = 'red')
sns.kdeplot(df_atiku['user_favourites'], label = 'Atiku', shade = True, color = 'blue')
sns.kdeplot(df_tinubu['user_favourites'], label = 'Tinubu', shade = True, color = 'yellow')


plt.title('Distributions of likes', size = 15)
plt.legend(prop={'size': 14})
plt.show()
```

```python
tinubu_nan = pd.Series(df_tinubu.isna().sum()[df_tinubu.isna().sum() > 0].
                      sort_values(ascending = False))
obi_nan = pd.Series(df_obi.isna().sum()[df_obi.isna().sum() > 0].
                      sort_values(ascending = False))
atiku_nan = pd.Series(df_atiku.isna().sum()[df_atiku.isna().sum() > 0].
                      sort_values(ascending = False))

fig, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(17, 5))
sns.set_style("whitegrid")
fig.suptitle('NaN values', size = 15)

sns.barplot(y = tinubu_nan.index, x = [len(df_tinubu)] * len(tinubu_nan),
            edgecolor = 'black', color = 'white', alpha = 0.6, ax = ax1)
sns.barplot(y = tinubu_nan.index, x = tinubu_nan, 
            edgecolor = 'black', alpha = 0.8, ax = ax1,
            palette = sns.color_palette("viridis", len(tinubu_nan)))
ax1.get_xaxis().get_major_formatter().set_scientific(False)
ax1.set_title('Tinubu dataset', size = 13)


sns.barplot(y = obi_nan.index, x = [len(df_obi)] * len(obi_nan),
            edgecolor = 'black', color = 'white', alpha = 0.6, ax = ax2)
sns.barplot(y = obi_nan.index, x = obi_nan, 
            edgecolor = 'black', alpha = 0.8, ax = ax2,
            palette = sns.color_palette("viridis", len(obi_nan)))
ax2.get_xaxis().get_major_formatter().set_scientific(False)
ax2.set_title('Obi dataset', size = 13)

sns.barplot(y = atiku_nan.index, x = [len(df_atiku)] * len(atiku_nan),
            edgecolor = 'black', color = 'white', alpha = 0.6, ax = ax3)
sns.barplot(y = atiku_nan.index, x = atiku_nan, 
            edgecolor = 'black', alpha = 0.8, ax = ax3,
            palette = sns.color_palette("viridis", len(atiku_nan)))
ax3.get_xaxis().get_major_formatter().set_scientific(False)
ax3.set_title('Atiku dataset', size = 13)


sns.despine()
```

```python
# Tinubu tweets authors
tinubu_count_tweets = pd.DataFrame(df_tinubu['user_name'].value_counts())
tinubu_count_tweets = pd.DataFrame({'user_name': tinubu_count_tweets.index, 
                            'count': tinubu_count_tweets.user_name})
tinubu_likes = df_tinubu[['user_name', 'favorites']].groupby('user_name').sum()
tinubu_agg = pd.merge(tinubu_count_tweets, tinubu_likes, on = 'user_name', 
                     how = 'left')

# obi tweets authors
obi_count_tweets = pd.DataFrame(df_obi['user_name'].value_counts())
obi_count_tweets = pd.DataFrame({'user_name': obi_count_tweets.index, 
                            'count': obi_count_tweets.user_name})
obi_likes = df_tinubu[['user_name', 'favorites']].groupby('user_name').sum()
obi_agg = pd.merge(obi_count_tweets, obi_likes, on = 'user_name', 
                     how = 'left')

# obi tweets authors
atiku_count_tweets = pd.DataFrame(df_atiku['user_name'].value_counts())
atiku_count_tweets = pd.DataFrame({'user_name': atiku_count_tweets.index, 
                            'count': atiku_count_tweets.user_name})
atiku_likes = df_tinubu[['user_name', 'favorites']].groupby('user_name').sum()
atiku_agg = pd.merge(atiku_count_tweets, atiku_likes, on = 'user_name', 
                     how = 'left')
```

```python
# The dependence of likes sums on tweets amounts
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5))
sns.set_style("whitegrid")
fig.suptitle("The dependence of likes sums on tweets amounts", size = 15)

sns.scatterplot(x = tinubu_agg['count'], y = tinubu_agg['favorites'],
                color = 'red', ax = ax1)
ax1.get_yaxis().get_major_formatter().set_scientific(False)
ax1.set_xlabel('Amount of tweets')
ax1.set_ylabel('Sum of likes')
ax1.set_title('Tinubu', size = 13)


sns.scatterplot(x = obi_agg['count'], y = obi_agg['favorites'],
                color = 'blue', ax = ax2)
ax2.get_yaxis().get_major_formatter().set_scientific(False)
ax2.set_xlabel('Amount of tweets')
ax2.set_ylabel('Sum of likes')
ax2.set_title('Obi', size = 13)

sns.scatterplot(x = atiku_agg['count'], y = atiku_agg['favorites'],
                color = 'yellow', ax = ax3)
ax3.get_yaxis().get_major_formatter().set_scientific(False)
ax3.set_xlabel('Amount of tweets')
ax3.set_ylabel('Sum of likes')
ax3.set_title('Atiku', size = 13)

fig.show()
```
